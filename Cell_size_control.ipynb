Cell size control can be modelled by assuming that, between birth and cell division, a cell's added size is given by $a s_b + \Delta$, where $a$ can take any value between $-1$ and $1$, $s_b$ is the size at birth and $\Delta$ is a fixed added size. 

In this notebook, I plot (blue line) the cell size at birth over several generations, given that the initial cell size at birth is $s_B = 20$. I also plot the steady state limit (orange line). This allows you to see how different values of $a$ affect the stability of cell size. 

```python
import matplotlib.pyplot as plt
import numpy as np


N_generations = 20
N = np.arange(N_generations)

delta = 5

sb = np.zeros(N_generations)
sb[0] = 20

fig, ((ax1, ax2, ax3),(ax4, ax5, ax6)) = plt.subplots(2,3, figsize = (20, 15))

a = 0.01
for i in range(1, N_generations):
  sb[i] = 0.5 * ((a+1)*sb[i-1] + delta)

limit = np.full(N_generations, delta/(1-a))

ax1.plot(N, sb)
ax1.plot(N, limit)

ax1.set(title = 'a=0.01',
        xlabel = 'Number of generations',
        ylabel = 'Cell size at birth')

a = 0.5
for i in range(1, N_generations):
  sb[i] = 0.5 * ((a+1)*sb[i-1] + delta)

limit = np.full(N_generations, delta/(1-a))

ax2.plot(N, sb)
ax2.plot(N, limit)

ax2.set(title = 'a=0.5',
        xlabel = 'Number of generations',
        ylabel = 'Cell size at birth')



a = 0.99
for i in range(1, N_generations):
  sb[i] = 0.5 * ((a+1)*sb[i-1] + delta)

limit = np.full(N_generations, delta/(1-a))

ax3.plot(N, sb)
ax3.plot(N, limit)

ax3.set(title = 'a=0.99',
        xlabel = 'Number of generations',
        ylabel = 'Cell size at birth')


a = -0.01
for i in range(1, N_generations):
  sb[i] = 0.5 * ((a+1)*sb[i-1] + delta)

limit = np.full(N_generations, delta/(1-a))

ax4.plot(N, sb)
ax4.plot(N, limit)

ax4.set(title = 'a=-0.01',
        xlabel = 'Number of generations',
        ylabel = 'Cell size at birth')

a = -0.5
for i in range(1, N_generations):
  sb[i] = 0.5 * ((a+1)*sb[i-1] + delta)

limit = np.full(N_generations, delta/(1-a))

ax5.plot(N, sb)
ax5.plot(N, limit)

ax5.set(title = 'a=-0.5',
        xlabel = 'Number of generations',
        ylabel = 'Cell size at birth')

a = -0.99
for i in range(1, N_generations):
  sb[i] = 0.5 * ((a+1)*sb[i-1] + delta)

limit = np.full(N_generations, delta/(1-a))

ax6.plot(N, sb)
ax6.plot(N, limit)

ax6.set(title = 'a=-0.99',
        xlabel = 'Number of generations',
        ylabel = 'Cell size at birth')

plt.show()


```
