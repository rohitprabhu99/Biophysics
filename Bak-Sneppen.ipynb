I evaluate the Bak-Sneppen model for a system with 100 species 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100 #Number of species
num_steps = 500 #Number of runs of simulation

B = np.random.random(N) #Array of B values (between 0 and 1)
B_min = np.zeros(num_steps) #Array of B_min (smallest B) values
B_min_max = np.zeros(num_steps) #Array of maximum B_min values

for k in range(num_steps):
  i_min = np.argmin(B) #Index of B_min

  #Storing and changing the B_min value
  B_min[k] = B[i_min]
  B_min_max[k] = max(B_min)
  B[i_min] = np.random.random()

  #Deciding which neighbour to change
  if i_min == N - 1:
    i_neighbour = N -2

  elif i_min == 0:
    i_neighbour = 1
  
  else:
    i_neighbour = i_min + np.random.choice(np.array([-1, 1]), p = [0.5, 0.5])

  #Changing the B value of the neighbour
  B[i_neighbour] = np.random.random()


#Plotting
plt.plot(B_min)
plt.plot(B_min_max)




```
